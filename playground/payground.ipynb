```python
import json
import re

def extract_correct_answers(text):
    # Find the "Tacni odgovori" section
    match = re.search(r'Tacni odgovori:\s*({.*})', text, re.DOTALL)
    if not match:
        return None
    
    json_data = match.group(1).strip()
    
    try:
        # Parse the JSON data
        correct_answers = json.loads(json_data)
        return correct_answers
    except json.JSONDecodeError:
        return None

# Example usage
# text = """
# ... (some text) ...
# Tacni odgovori
# {
#     "question id": "1",
#     "correct answer": "A"
# }
# ... (more text) ...
# """

correct_answers = extract_correct_answers(quiz)
if correct_answers:
    print(correct_answers)
else:
    print("No valid 'Tacni odgovori' section found.")
```
